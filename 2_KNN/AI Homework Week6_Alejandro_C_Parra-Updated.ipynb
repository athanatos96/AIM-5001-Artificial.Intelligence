```python
import numpy as np

from scipy.io import loadmat
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
```

# 1. Implement k-NN method

```python
# calculate disntance between points. 
def euclidean_distance(point_a, point_b):
    sum = 0
    for elem_a, elem_b in zip(point_a,point_b):
        sum += (elem_a-elem_b)**2
    return np.sqrt(sum)

def get_closest_k_points(new_point, points_list, k=3):
    distances = []
    for elem in points_list:
        distance = euclidean_distance(new_point, elem)
        distances.append(distance) 
    #return np.argpartition(distances, k)[:k] # get the position of the closest k elements
    
    return (distances, np.argsort(distances)[:k])# get the position of the closest k elements

def knn_method(test_points, train_points, train_labels, k=3):
    count = 0
    predicted_labels = []
    for new_point in test_points:
        if (count%100 == 0):
            print(count)
        count+=1
        # get closest k elements
        _ , closest_elements = get_closest_k_points(new_point, train_points, k=k)
        #print(closest_elements)
        # get the lables of those k-closest elements
        label = []
        for closest_elem in closest_elements:
            label.append(train_labels[closest_elem][0])
        #prediction = np.bincount(label).argmax()# get most frequent value
        #print(label)
        prediction = max(set(label), key=label.count)# get most frequent value
        #print(prediction)
        predicted_labels.append(prediction)# save prediction
        #print()
    return np.array(predicted_labels)
```

# 2. Load train and test mat files, perform k-NN and report acuracy on the test dataset

```python

# Open the model
train_dataset = loadmat('train_updated.mat')
#print(annots)
```

```python
train_dataset
```

```python
train_features = train_dataset['features']
train_labels = train_dataset['labels']
print(train_features.shape)
print(train_labels.shape)
```

```python

# Open the model
test_dataset = loadmat('test_updated.mat')
#print(annots)
```

```python
test_dataset
```

```python
test_features = test_dataset['features']
test_labels = test_dataset['labels']
print(test_features.shape)
print(test_labels.shape)
```

```python
# call the method
predictions = knn_method(test_features, train_features, train_labels, k=5)
predictions
```

```python
np.moveaxis(test_labels, 0, -1)[0]
```

```python
# Calculate Accuracy
acc = accuracy_score(np.moveaxis(test_labels, 0, -1)[0], predictions)

print("Accuracy: ", acc*100, "%")
```

```python
pca = PCA(n_components=10)
pca.fit(train_features)
```
